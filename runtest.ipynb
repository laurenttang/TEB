```python
from ckiptagger import WS, POS, NER
ws = WS('/Users/riversoft/Desktop/TEB/專題/ckipdata')
#pos = POS(<path>)
#ner = NER(<path>)
```

```python
import pandas as pd
word_example = pd.read_csv(r'/Users/riversoft/Desktop/TEB/專題/104.csv')
t1=word_example.loc[0,'jobDescription']
print('\n--head_columns--\n',word_example.columns)
print('\n--choosed_columns--\n',word_example['jobName'])
print('\n--choosed_desc--\n',''.join(t1))
```

```python
sentence_list=[]
for i in word_example['jobDescription']:
    sentence_list.append(i)
print(sentence_list,len(sentence_list))

word_s = ws(sentence_list,
            sentence_segmentation=True,
            segment_delimiter_set={'?', '？', '!', '！', '。', ',',   
                                '，', ';', ':', '、'}
           )
print(word_s)
```
